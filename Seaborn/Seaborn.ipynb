```python
cr=dl[['goals','fouls']]
cr
```

```python
cr.corr(method='pearson')
```

```python
sns.heatmap(data =cr.corr(method='pearson'),vmin =-1,vmax =1,annot =True,cmap='crest')
```

```python
sns.heatmap(data =cr.corr(method='spearman'),vmin =-1,vmax =1,annot =True,cmap='crest')
```
