# Subplots

```python
co=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]

plt.figure(figsize=(7,6))

plt.subplot(2,2,1)
plt.plot(co,a,label = "Co_A",color = 'r')
plt.plot(co,b,label="Co_B")
plt.plot(co,c,label="Co_c",color = 'g')
plt.legend()
plt.title('3 combination')

plt.subplot(2,2,2)
plt.plot(co,a,color = 'r')
plt.title('Co_A',)


plt.subplot(2,2,3)
plt.plot(co,b)
plt.title('Co_B')

plt.subplot(2,2,4)
plt.plot(co,c,color = 'g')
plt.title('Co_C')
plt.show()
```

# Barchart

```python
course = ['Python','Java','.NET c#','GO']
revenue= [900,1660,891,247]
```

```python
len(course)
```

```python
a = np.arange(len(course))
a
```

```python
plt.figure(figsize=(5,3))
categories = ['a' , 'b' , 'c' , 'd']
values = np.arange(1,5)

plt.bar(categories,values)
plt.show()
```

```python
plt.figure(figsize=(5,3))
categories = ['a' , 'b' , 'c' , 'd']
values = np.array([25,50,75,30])

plt.bar(categories,values)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
cities = np.array(['New York', 'Tokyo', 'London', 'Paris', 'Sydney'])
population = np.array([8623000, 9273000, 8908081, 2140526, 5312000])
colors = np.array(['blue', 'orange', 'green', 'red', 'purple'])

plt.bar(cities,population,color = colors)
plt.ylabel('count')
plt.xlabel('cities')
plt.title("population of cites")
plt.show()
```

```python
plt.figure(figsize=(5,3))
cities = np.array(['New York', 'Tokyo', 'London', 'Paris', 'Sydney'])
population = np.array([8623000, 9273000, 8908081, 2140526, 5312000])
colors = np.array(['blue', 'orange', 'green', 'red', 'purple'])

plt.bar(cities,population,color = colors)
plt.grid(True)
plt.ylabel('count')
plt.xlabel('cities')
plt.title("population of cites")
plt.show()
```

# piechart

```python
plt.figure(figsize=(7,5))

x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels = y)
```

```python
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels = y,shadow = True)
```

```python
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels = y,shadow = True,autopct = "%1.1f%%")
plt.show()
```

```python
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
explode = [0.0,0.6,0,0.0,0.0]
plt.pie(x,labels = y,shadow = True,autopct = "%1.1f%%",explode = explode)
plt.show()
```
